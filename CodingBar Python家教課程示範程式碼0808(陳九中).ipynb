英打比賽計分
--
Ivy 今天要與 Slime 比賽英打速度，
贏的人可以獲得獎金 10000 元，
比賽規則是只要打出一個單字就可以獲得 1 分
(單字指的是由 a ~ z 或 A ~ Z 組合而成的字 ) 。

- 例如："this is an apple .... but I can't eat"




上方句子總得分為 9 分。
現在請你設計一個評分系統，來計算一串文字的總分為多少。


```python
lower_letter = "abcdefghijklmnopqrstuvwxyz"
upper_letter = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
while True:
    n = input()
    if n == "0":
        break
    flag = False
    count = 0 
    for i in n:
        if i in lower_letter or i in upper_letter:
            if flag == False:
                count += 1
            flag = True
        else:
            flag = False
    print(count)
```

密碼強度檢驗
--
請讓使用者可以輸入一個文字，
用來表示設定的密碼，
透過系統協助檢查使用者設置的密碼是否符合規定？
密碼規定如下：
1. 密碼須皆由 數字 或 字母 或 特殊符號 組成
2. 密碼長度不能低於 8 位

請經過程式檢驗之後，依據不同情況分別顯示對應結果。


請注意條件應由上往下一一檢驗，檢查完前一項之後繼續檢查下一項條件。

範例輸入 / 輸出：
--



```python
#建議特殊符號檢查表。

#symbols = "!@#$%^&*()_+-=/{}[]\|;:/?,.<>"
#chars = "abcdefghijklmnopqrstuvwxyz"
#chars_b = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
#nums = "0123456789"
```

```python
symbols = "!@#$%^&*()_+-=/{}[]\|;:/?,.<>"
chars = "abcdefghijklmnopqrstuvwxyz"
chars_b = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
nums = "0123456789"

passwd = input() #提示：要讓使用者可以輸入資料
length = len(passwd)  #提示：取得輸入文字長度
flag = 1  #提示：此題希望利用 flag 儲存不同數字，用來表示不同狀態(或層級！)

#第一條件 - 長度不低於 8
if length > 8:
    #第二條件 - 迴圈迭代找出是否皆由數字或文字所組成
    for i in passwd:
        if i in symbols or i in chars or i in chars_b or i in nums:
            flag = 1
        else:
            flag = 0
            break
else:
    #長度確實不到 8，一樣直接表示最低 False 程度
    flag = 0

if flag == 0:
    print(False)
else:
    flag_con = 0
    # 判斷是否包含符號
    for each in passwd:
        if each in symbols:
            flag_con += 1
            break
    # 判斷是否包含字母
    for each in passwd:
        if each in chars or chars_b:
            flag_con += 1
            break
    # 判斷是否包含數字
    for each in passwd:
        if each in nums:
            flag_con += 1
            break

    if flag_con != 3:
        print("低")
    else:
        if passwd[0] in chars_b:
            print("高")
        else:
            print("中")
```

布林表
--


```python
n = 50
table = [False for i in range(n+1)]
for number in range(2,n+1):
    for i in range(2,number):
        if number % i == 0:
            break
    else:
        table[number] = True
print(table)
print(len(table))
```

英文拼字表
--
有很多英文單字都喜歡重複使用英文字，例如：Mississippi 密西西比州在這個單字當中，就重複出現了好幾次的 i 和 s，當然你也可以說有 p，是不是看了頭很痛呢？現在如果給你一個英文單字，你能不能判斷單字當中，是否有重複的英文字母呢？

注意：只要判斷重複即可，不用考慮連續的問題！



```python
text = input()

repeat = []
for i in text:
    if i in repeat:
        print(True)
        break
    else:
        repeat.append(i)
else:
    print(False)
```

```python
#另解
text = input()

letter = "abcdefghijklmnopqrstuvwxyz"
boolean_table = [False for i in range(26)]

for i in text:
    index = letter.index(i)  #第幾個英文字母在第幾個index
    if boolean_table[index] == True: #先檢查那一格是不是True
        print(True)
        break
    else:
        boolean_table[index] = True
else:
    print(False)
```
