投票統計
--
請設計一個程式，讓使用者重複輸入數字，
代表記錄投票的結果，
當輸入 no 的時候輸入結束，
並顯示出哪個數字得票超過一半。

```python
a = []          #為什麼a要放在這一行，不放在迴圈裡面？
while True:
    num = input()
    if num == "no":
        break
    else:
        num = int(num)
        a.append(num)
set_a = set(a)    #建立Set (利用Set的特性，確認有多少不重複的元素)
print(set_a)
b = []            #建立另一空串列b，用來計算每個元素共出現了幾次。
for i in (set_a):
    count = 0 
    for j in a:
        if i == j:
            count += 1
    b.append(count)
print(b)
for k in range(len(b)):
    if b[k] >= int(len(a)/2+1):
        print(list(set_a)[k],"超過了一半，共出現了" ,b[k] ,"次")
        break
    elif k == len(b)-1:
        print("wrong, 沒有任何數過半")
        break        
```

```python
number_list = []
count_list = []
while True:
    number = input()
    if number=="no":
        break
    else:
        number = int(number)
        number_list.append(number)

set_number = set(number_list)
for i in (set_number) :
    count = number_list.count(i)
    count_list.append(count)
for j in range (len(count_list)):
    if count_list[j] >= int(len(number_list)/2+1):
        print(list(set_number)[j])
        break

        
print(len(number_list))
print(set_number)
print(count_list)
```

#INPUT

輸入共有 1 + n 列，第 1 列必為一個正整數 n，

找方程式解
--
小明今天在上數學課的時候，看到一題方程式的題目：
給你四個未知數 a, b, c, d，請找出 a + b + c + d = 20 所有符合的可能性，
" 其中有個限制要求！就是 a < b < c < d 且 每個未知數都是 1 ~ 9 的數字。 "

小明想了想，覺得這題很簡單，
因為只要讓 a = 1, b = 2, c = 3, d = 4，然後加加看有沒有符合 20 就好，
如果沒有就再換一組數字，a = 1, b = 2, c = 3, d = 5，依此類推，
直到找到為止，雖然會花很多時間，但一定可以找到答案。

為了要讓小明不能太快看到答案，記得程式要設計先讓小明輸入一個文字，
輸入任何文字時都不顯示東西，只有當小明輸入 "answer" 的時候，
才會將所有符合方程式 a + b + c + d = 20, a < b < c < d 的答案列出來!



- #INPUT
- 輸入含有一個文字，用來表示顯示答案的通關密語。



- #OUTPUT
- 輸出共有多列，皆含有四個整數，用來表示所有符合方程式的解，

中間請用一格空格分隔。

四個整數的順序應為 a, b, c, d，且 a < b < c < d。

不同組答案的輸出順序應從 a 最小開始，當 a 相同時則應從 b 最小開始，依此類推。

```python

while True:
    key = input("請輸入密碼")
    if key == "answer":
        for a in range(1,10):
            for b in range(2,10):
                for c in range(3,10):
                    for d in range(4,10):
                        if a+b+c+d == 20 and a<b<c<d:
                            print(a,b,c,d)      
        break
```

可以自行定義密碼的版本如下：

```python
ans = input("請輸入通關密碼")
while True:
    key = input("請輸入密碼")
    if key == ans:
        for a in range(1,10):
            for b in range(2,10):
                for c in range(3,10):
                    for d in range(4,10):
                        if a+b+c+d == 20 and a<b<c<d:
                            print(a,b,c,d)
                            
        break
```

成雙成對
--


在撲克牌的遊戲當中，不乏衍生出各種類型的玩法，其中很常透過牌與牌之間的組合進行比大小，而最簡單湊成的牌型就是一對(pair)。

現在請你設計一個程式，讓使用者可以重複輸入一張撲克牌的類型，
請試著判別是否能湊成一對，並將湊成的牌型顯示出來，
當輸入 "no" 的時候，程式結束，
並將剩下未湊成一對的牌依照輸入順序依序顯示。

#INPUT

輸入共有多列，皆為一個文字，用來表示撲克牌，

文字格式必為一個 "s", "d", "h", "c" 其中的字配上一個 1~9 的數字。

(spade, diamond, heart, club)

#OUTPUT

輸出共有多列，皆為一個文字，用來表示成對的兩張撲克牌，

最後一列為一個串列，用來表示收集到的剩下未成對撲克牌。



```python
a = []    #用a串列把
while True:
    poke = input("請輸入撲克花色與數字")
    if poke == "no":
        break
    for index,i in enumerate(a):    #一個迭代索引值、另一個迭代串列裡面的值
        if poke[1] == i[1]:
            print(i, poke)
            del a[index]
            break
    else:
        a.append(poke)
            
print(a)
        
```

不吉利的數字
--
請設計讓使用者輸入一串數字，
利用 for迴圈 將該數字的每一位數字檢查，
如果找到有出現 4，那麼就顯示 4。

當輸入數字時輸入 "no"，程式結束。



```python

while True:
    num = input("請輸入數字")
    if num == "no":
        break   
    for index, i in enumerate(num):
        if i =="4":
            print(4)
        
```

將所有位數的數字收集起來！
--

請設計讓使用者可以重複輸入一串數字，
利用 for迴圈 將該數字的每一位數字進行加總，
並將結果顯示出來。
當輸入數字時輸入 "no"，程式結束。



```python

while True:
    count = 0
    num = input("請輸入數字")
    if num == "no":
        break   
    for index, i in enumerate(num):
        count += int(i)
    print(count)
        
        
        
```

請設計讓使用者可以重複輸入一個文字，
利用程式判斷輸入的每一個字是不是都是阿拉伯數字，
如果有找到任何一個不是的情況，就顯示 "False"，
